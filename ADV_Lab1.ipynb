```python
import pandas as pd

df = pd.read_csv("myntra_products_catalog.csv")
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df["Gender"], kde=True)
plt.show()
```

```python
brand_data = df[df["ProductBrand"] == "Raymond"]

product_sales = brand_data.groupby("ProductID")["Sales"].sum().reset_index()

top_product_sales = product_sales.sort_values(by='Sales', ascending=False).head(5)

sns.barplot(x='ProductID', y='Sales', data=top_product_sales)

plt.title('Product-wise Total Sales for Raymond')
plt.xlabel('Product ID')
plt.ylabel('Total Sales')

plt.show()
```

```python
color_sales = df.groupby('PrimaryColor')['Sales'].sum().reset_index()

top_5_colors = color_sales.sort_values(by='Sales', ascending=False).head(5)

plt.figure(figsize=(8, 8))
plt.pie(top_5_colors['Sales'], labels=top_5_colors['PrimaryColor'], autopct='%1.1f%%', colors=sns.color_palette('pastel', n_colors=5))

plt.title('Top 5 Primary Color-wise Sales Distribution')

plt.show()
```

```python
brand_data = df[df['ProductBrand'] == 'Parx']

plt.figure(figsize=(10, 6))

sns.scatterplot(
    x='ProductID',
    y='Price (INR)',
    size='Sales',
    hue='ProductID',
    data=brand_data,
    palette='viridis',
    sizes=(50, 400),
    legend=False
)

plt.title('Price Range of Parx Products')
plt.xlabel('ProductID')
plt.ylabel('Price')

plt.show()
```
